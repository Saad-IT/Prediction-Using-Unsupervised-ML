### Problem Statement: From the given ‘Iris’ dataset, predict the optimum number of  clusters and represent it visually.

#### Importing the required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D      
from matplotlib import cm
```

#### Loading the iris dataset

```python
iris = datasets.load_iris()
```

#### Converting data into data frame format

```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head() # See the first 5 rows
```

### KMeans

#### Finding optimum number of clusters

```python
x = iris_df.iloc[:, [0, 1, 2, 3]].values


from sklearn.cluster import KMeans
wcss = [] #the empty list will contain Sum of squared distances of samples to their closest cluster center

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
# Plotting the results onto a line graph, 
# allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

#### You can clearly see why it is called 'The elbow method' from the above graph, the optimum clusters is where the elbow occurs. This is when the within cluster sum of squares (WCSS) doesn't decrease significantly with every iteration.


#### From this we choose the number of clusters as '3'.

### Creating the kmeans classifier / Applying kmeans to the dataset

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

### Visualising the clusters - On the first three columns

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'black', label = 'Centroids')
plt.grid(True)

plt.legend()
```

### 3D Visualisation

```python
fig=plt.figure(figsize=(20,20))
ax=fig.add_subplot(projection='3d')

#Cluster 1
ax.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], x[y_kmeans == 0, 2],
            s = 150, c = 'red', label = 'Iris-setosa')

#Cluster 2
ax.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], x[y_kmeans == 1, 2],
            s = 150, c = 'green', label = 'Iris-versicolour')

#Cluster 3
ax.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], x[y_kmeans == 2, 2],
            s = 150, c = 'blue', label = 'Iris-virginica')

# Plotting the centroids of the clusters
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],kmeans.cluster_centers_[:,2], 
            s = 150, c = 'black', label = 'Centroids')
```
